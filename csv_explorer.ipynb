```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the sine wave samples from the CSV file
csv_path = '/home/thanostriantafyllou/GS4Time/data/time_series/chair/sine_wave_samples.csv'
df = pd.read_csv(csv_path, index_col='period')

# Display the DataFrame
df.head()
```

```python
# Concatenate all the samples into a single sequence
whole_signal = df.values.flatten()

# Generate a time index for the whole signal
time_index = np.arange(len(whole_signal))

# Number of samples per window
samples_per_period = df.shape[1]

# Use a colormap
cmap = plt.get_cmap('tab20')
colors = cmap(np.linspace(0, 1, df.shape[0]))

# Plot the whole sine wave signal with colored windows
plt.figure(figsize=(12, 6))

# Plot the entire signal
plt.plot(time_index, whole_signal, color='black')
plt.scatter(time_index, whole_signal, color='black', s=5)

# Shade the background for each window
for i in range(df.shape[0]):
    start_idx = i * samples_per_period
    end_idx = (i + 1) * samples_per_period
    plt.axvspan(start_idx, end_idx, facecolor=colors[i], alpha=0.3)

plt.xlabel('Sample Index')
plt.ylabel('Sine Value')
plt.title('Whole Sine Wave Signal with Colored Windows')
plt.grid(True)
plt.show()
```
